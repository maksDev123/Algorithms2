### Proof by Mathematical Expectation

Let N_1, N_2, ... , N_k be the subproblems we have at the k-th step.

$N_i$ is a successful step if $|N_i| < \frac{5}{6} |N_{i-1}|$.

$Y_i$ - the number of recursive calls after the $(i-1)$-th successful iteration,

$n_i$ - the size of the subproblem after the $(i-1)$-th successful iteration.

The execution time of the algorithm is $T - \sum_{k=0}^{i} n_k \cdot Y_k$.

Let $n_i < \frac{5}{6}n_{i-1} < \frac{5}{6}n$.

$Y_i$ is a random variable with a geometric distribution with a parameter of $\frac{1}{2}$.

$E(Y_i) = 2$.
$E\left(\sum_{k=0}^{i} n_k \cdot Y_k\right) = \sum_{k=0}^{i} \frac{5}{6}n \cdot Y_k = n\sum_{k=0}^{i}\frac{5}{6} E(Y_k) < n\sum_{k=0}^{i} \frac{5}{6} \cdot 2$.

Where a is some constant. This means that the expected time complexity of quickSelect is linear.

### Proof by Mathematical Induction

* ##### Base: 
    
    Let n be the number of elements in the list.

    For  n = 1, then the execution time T(1) = 0.


* ##### Assumption: 

    Let's assume $T(t) \leq a \cdot t$ as our inductive hypothesis for t < n.


* ##### Inductive step: 
    
    $T(n) < (time_to_split) + (time_for_subproblem) + (partition_time)$
    
    Time to split - (O(1))

    Partition time = O(n)

    Time for subproblem O(linear of n) 


This gives that:
$T(n) \leq O(n) + O(1) + 2 \cdot O(n)$. ---> $T(n) = O(n)$.

3.2 a) 
The best case for QuickSort is when pivot devides array directly on two parts of the same size - time complexity - O(n * log n)

`T(n) = 2 * T(n/2) + O(n) -> T(n) = O(n log n)`

The worst case is when pivot get first or last element in sorted order - time complexity - O(n^2)

`T(n) = T(n - 1) + O(n) -> T(n) = O(n^2)`

3.2) b)

The best case for determenistic median finding is when this deterministic pivot points on exact median, so for this time complexity will be only O(n) (partition)


The worst case is when pivot of determenistic median finding points on last or first in sorted order element, decreasing problem just by one, in this case complexity is O(n^2)

It can be shown by this formula:

`T(n) = T(n - 1) + O(n) -> T(n) = O(n^2)`
